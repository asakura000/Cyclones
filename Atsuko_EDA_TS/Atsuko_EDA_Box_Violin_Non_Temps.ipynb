```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

sns.set()

%matplotlib inline

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.columns
```

### Categorical plots: 
Purpose: To see distribution of a categorical column in reference to a numerical column or another categorical column

## List of Plots in this Notebook:  
1. Box Plot:
    1) Variable only
2. Violin Plot:
    1) Variable only

```python
def boxplots(col):
    
    fig, axes = plt.subplots(1, 2, figsize=(30, 10))
    fig.suptitle(col)
    axes[0].set_title('Train')
    axes[1].set_title('Test')
    sns.boxplot(ax=axes[0], x=col, data=train, color='aqua')
    sns.boxplot(ax=axes[1], x=col, data=test, color='pink')
    
```

```python
def violinplots(col):
    
    fig, axes = plt.subplots(1, 2, figsize=(30, 10))
    fig.suptitle(col)
    axes[0].set_title('Train')
    axes[1].set_title('Test')
    sns.violinplot(ax=axes[0], x=col, data=train, color='aqua')
    sns.violinplot(ax=axes[1], x=col, data=test, color='pink')
    
```

## 6.1 Year Built Box Plots

```python
con1 = train['year_built'] == 0
con2 = train['year_built'].isna()
has_year_train = train[~con1 & ~con2]

con1 = test['year_built'] == 0
con2 = test['year_built'].isna()
has_year_test = test[~con1 & ~con2]
```

```python
col = 'year_built'
fig, axes = plt.subplots(1, 2, figsize=(30, 10))
fig.suptitle('Year Built')
axes[0].set_title('Train')
axes[1].set_title('Test')
sns.boxplot(ax=axes[0], x=col, data=has_year_train, color='aqua')
sns.boxplot(ax=axes[1], x=col, data=has_year_test, color='pink')
```

## 6.2 Year Built Violin Plots

```python
fig, axes = plt.subplots(1, 2, figsize=(30, 10))
fig.suptitle('Year Built')
axes[0].set_title('Train')
axes[1].set_title('Test')
sns.violinplot(ax=axes[0], x=col, data=has_year_train, color='aqua')
sns.violinplot(ax=axes[1], x=col, data=has_year_test, color='pink')
```
