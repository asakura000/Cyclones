```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

sns.set()

%matplotlib inline

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
pivot_energy_star = np.round(pd.pivot_table(train, values=['energy_star_rating'],
                                index=['facility_type', 'building_class'],
                                columns= ['State_Factor', 'Year_Factor'],
                                aggfunc=[np.median]), 2)
```

```python
pivot_site_eui = np.round(pd.pivot_table(train, values=['energy_star_rating'],
                                index=['facility_type', 'building_class'],
                                columns= ['State_Factor', 'Year_Factor'],
                                aggfunc=[np.median]), 2)
```

```python
pivot_energy_star.style.format({'site_eui': '${0:, .0f}',
                    'energy_star_rating': '{0:, .0f}'})
```

```python
pivot_site_eui.style.format({'site_eui': '${0:, .0f}',
                    'energy_star_rating': '{0:, .0f}'})
```

```python
plt.figure(figsize = (21,13))
sns.heatmap(pivot_energy_star, cmap='coolwarm', annot=True, linewidth=1, linecolor='black')
```

```python
plt.figure(figsize = (21,13))
sns.heatmap(pivot_site_eui, cmap='coolwarm', annot=True, linewidth=1, linecolor='black')
```
