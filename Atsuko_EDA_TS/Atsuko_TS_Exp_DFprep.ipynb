## Grab all facility types column names

```python
facs_list = list(train['facility_type'].unique())
```

```python
pivot_fac = np.round(pd.pivot_table(train, values=['site_eui'],
                                index=['Year_Factor'],
                                columns= ['facility_type'],
                                aggfunc=[np.mean]), 2)
```

```python
pivot_state = np.round(pd.pivot_table(train, values=['site_eui'],
                                index=['Year_Factor'],
                                columns= ['State_Factor'],
                                aggfunc=[np.mean]), 2)
#pivot_state.style.format({'site_eui': '${0:, .0f}'})
pivot_state
```

```python
fig, ax = plt.subplots(figsize=(21,13))       
sns.heatmap(pivot_fac.isnull(), yticklabels=False, cbar=False, cmap='vlag')
```

```python
nbr_nulls = pivot_fac.isnull().sum()
```

```python
null_cols = nbr_nulls[nbr_nulls > 0]
null_cols
```

```python
null_cols = null_cols.reset_index()
```

```python
list_nulls = null_cols['facility_type'].unique()
list_nulls
```

## **Set Columns in Pivot Table to be Facility Type**

Right now I have a multi-index column. This will make calculations for imputing data annoying so I want a dataframe with just the year as an index and facility types as columns. 

```python
pivot_fac
```

## Drop the top level index (mean)

```python
pivot_fac = pivot_fac.droplevel(level=0, axis=1)
```

## Drop the next level index (site_eui)

```python
pivot_fac = pivot_fac.droplevel(level=0, axis=1)
```

```python
pivot_fac
```

### See if it worked...

```python
pivot_fac.columns
```

### For each column from the original pivot that had null values, impute the cell with the mean for each column

```python
for col in list_nulls:
    mean_value = pivot_fac[col].mean()
    pivot_fac[col].fillna(value = mean_value, inplace=True)
```

```python
pivot_fac
```

## Check to make sure all nulls are gone

```python
fig, ax = plt.subplots(figsize=(21,13))       
sns.heatmap(pivot_fac.isnull(), yticklabels=False, cbar=False, cmap='coolwarm')
```
