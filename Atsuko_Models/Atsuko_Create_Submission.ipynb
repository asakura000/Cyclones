```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

sns.set()

%matplotlib inline

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

## TRAIN ONLY

### Feature Engineering 1: Drop rows where "year built" is either null or 0

```python
train = train[train['year_built'].notna()]
train = train[train['year_built'] != 0]
```

```python
train.shape
```

```python
# sanity check
# train['year_built'].unique()
```

### Feature Engineering 2: Drop these columns:
    - direction_max_wind_speed
    - direction_peak_wind_speed
    - max_wind_speed
    - days with fog

```python
train = train.drop(columns = ['direction_max_wind_speed', 'direction_peak_wind_speed', 'max_wind_speed', 'days_with_fog'])
```

```python
test = test.drop(columns = ['direction_max_wind_speed', 'direction_peak_wind_speed', 'max_wind_speed', 'days_with_fog'])
```

### Feature Engineering 3: Impute ESR with mean for each facility type

```python
# create a series 
gb_train = train.groupby('facility_type')['energy_star_rating'].mean()
gb_test = test.groupby('facility_type')['energy_star_rating'].mean()

# convert the series to a dictionary
dict_train = gb_train.to_dict()
dict_test = gb_test.to_dict()
```

```python
# sanity check - before 

print('training nulls:', train['energy_star_rating'].isnull().sum())
print('testing nulls:', test['energy_star_rating'].isnull().sum())
```

```python
# fill all null values in the energy_star_rating column with corresponding values mapped from the dictionary 

train['energy_star_rating'] = train['energy_star_rating'].fillna(train['facility_type'].map(dict_train))
test['energy_star_rating'] = test['energy_star_rating'].fillna(test['facility_type'].map(dict_test))
```

```python
# sanity check - after 

print('training nulls:', train['energy_star_rating'].isnull().sum())
print('testing nulls:', test['energy_star_rating'].isnull().sum())
```

```python
# now fill the remaining null values for energy star rating with the overall mean

train_esr_mean = train['energy_star_rating'].mean()
test_esr_mean = test['energy_star_rating'].mean()

train['energy_star_rating'].fillna(value = train_esr_mean, inplace=True)
test['energy_star_rating'].fillna(value = test_esr_mean, inplace=True)
```

```python
# final sanity check 

print('training nulls:', train['energy_star_rating'].isnull().sum())
print('testing nulls:', test['energy_star_rating'].isnull().sum())
```

### Feature Engineering 4: Dummy encode State_Factor, business_class,  facility_type

```python
train = pd.get_dummies(train, columns=['State_Factor', 'building_class', 'facility_type'], drop_first=False)
test = pd.get_dummies(test, columns=['State_Factor', 'building_class', 'facility_type'], drop_first=False)
```

```python
print('train:', train.shape)
print('test: ', test.shape)
```

```python
list_tr = list(train.columns)
list_te = list(test.columns)

def Diff(li1, li2):
    return list(set(li1) - set(li2)) + list(set(li2) - set(li1))

Diff(list_tr, list_te)
```

### Define X and Y 

```python
# training set

all_cols_tr = list(train.columns)
remove = ['site_eui', 'id']
features = [i for i in all_cols_tr if i not in remove]
target = ['site_eui']

X_train = train[features]
y_train = train[target]
```

```python
# testing set

all_cols_te = list(test.columns)
remove = ['site_eui', 'id']
features = [i for i in all_cols_te if i not in remove]
target = ['site_eui']

X_test = test[features]
X_test['State_Factor_State_6'] = 0
```

### XGBoost Regressor

```python
import xgboost as xgb
```

```python
params={ 'objective':'reg:squarederror',
         'max_depth': 6, 
         'colsample_bylevel':0.5,
         'learning_rate':0.01,
         'random_state':20}
```

```python
xgb_r = xgb.XGBRegressor(params=params)
```

```python
xgb_r.fit(X_train, y_train)
```

```python
test['site_eui'] = xgb_r.predict(X_test)
test = test[['id', 'site_eui']]
```
