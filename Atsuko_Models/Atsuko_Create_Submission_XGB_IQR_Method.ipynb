## Remove Outliers Method 2: IQR Method

```python
percentile25 = train['site_eui'].quantile(0.25)
percentile75 = train['site_eui'].quantile(0.75)
iqr = percentile75 - percentile25
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
len(train[train['site_eui'] > upper_limit])
#len(train[train['site_eui'] < lower_limit])
```

```python
new_train = train[train['site_eui'] < upper_limit]
new_train.shape
```

```python
new_train['site_eui'].describe()
```

```python
plt.figure(figsize=(13,8))
sns.distplot(new_train['site_eui'], color='red')
```

```python
train2 = new_train[train['year_built'].notna()]
train2 = new_train[train['year_built'] != 0]
```

### Feature Engineering 2: Drop these columns:
    - direction_max_wind_speed
    - direction_peak_wind_speed
    - max_wind_speed
    - days with fog

```python
train2 = new_train.drop(columns = ['direction_max_wind_speed', 'direction_peak_wind_speed', 'max_wind_speed', 'days_with_fog'])
test = test.drop(columns = ['direction_max_wind_speed', 'direction_peak_wind_speed', 'max_wind_speed', 'days_with_fog'])
```

### Feature Engineering 3: Impute ESR with mean for each facility type

```python
# create a series 
gb_train = train2.groupby('facility_type')['energy_star_rating'].mean()
gb_test = test.groupby('facility_type')['energy_star_rating'].mean()

# convert the series to a dictionary
dict_train = gb_train.to_dict()
dict_test = gb_test.to_dict()
```

```python
# fill all null values in the energy_star_rating column with corresponding values mapped from the dictionary 

train2['energy_star_rating'] = train2['energy_star_rating'].fillna(train['facility_type'].map(dict_train))
test['energy_star_rating'] = test['energy_star_rating'].fillna(test['facility_type'].map(dict_test))
```

```python
# now fill the remaining null values for energy star rating with the overall mean

train_esr_mean = train2['energy_star_rating'].mean()
test_esr_mean = test['energy_star_rating'].mean()

train2['energy_star_rating'].fillna(value = train_esr_mean, inplace=True)
test['energy_star_rating'].fillna(value = test_esr_mean, inplace=True)
```

```python
# final sanity check 

print('training nulls:', train2['energy_star_rating'].isnull().sum())
print('testing nulls:', test['energy_star_rating'].isnull().sum())
```

## Dummy encode categories

```python
train2 = pd.get_dummies(train2, columns=['State_Factor', 'building_class', 'facility_type'], drop_first=False)
test = pd.get_dummies(test, columns=['State_Factor', 'building_class', 'facility_type'], drop_first=False)
```

```python
print('train:', train2.shape)
print('test: ', test.shape)
```

```python
list_tr = list(train2.columns)
list_te = list(test.columns)

def Diff(li1, li2):
    return list(set(li1) - set(li2)) + list(set(li2) - set(li1))

Diff(list_tr, list_te)
```

### Define X and Y 

```python
# training set

all_cols_tr = list(train2.columns)
remove = ['site_eui', 'id']
features = [i for i in all_cols_tr if i not in remove]
target = ['site_eui']

X_train = train2[features]
y_train = train2[target]
```

```python
# testing set

all_cols_te = list(test.columns)
remove = ['site_eui', 'id']
features = [i for i in all_cols_te if i not in remove]
target = ['site_eui']

X_test = test[features]
X_test['State_Factor_State_6'] = 0
```

### XGBoost Regressor

```python
import xgboost as xgb
```

```python
params={ 'objective':'reg:squarederror',
         'max_depth': 6, 
         'colsample_bylevel':0.5,
         'learning_rate':0.01,
         'random_state':20}
```

```python
xgb_r = xgb.XGBRegressor(params=params)
```

```python
xgb_r.fit(X_train, y_train)
```

```python
test['site_eui'] = xgb_r.predict(X_test)
test = test[['id', 'site_eui']]
```
